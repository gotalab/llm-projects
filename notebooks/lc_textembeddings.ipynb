### テキストの類似度の算出

```python
import pandas as pd

df = pd.DataFrame(data=["School", "College", "Car", "Bike", "Apple", "Orange", "Banana"], columns=["words"])
```

```python
df["embedding"] = df["words"].apply(lambda x: embeddings.embed_query(x))
df.to_csv("output/word_embeddings.csv")
```

```python
new_df = df.copy()
```

```python
text = "Mango"
```

```python
text_embedding = embeddings.embed_query(text)
```

```python
from openai.embeddings_utils import cosine_similarity

df['similarity_score'] = df['embedding'].apply(lambda x: cosine_similarity(x, text_embedding))
```

```python
df.sort_values("similarity_score", ascending=False)
```
